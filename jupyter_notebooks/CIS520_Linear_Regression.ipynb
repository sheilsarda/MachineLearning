```python
x = np.random.random(10) 
x = np.sort(x)
```

```python
X = np.vstack([np.ones(len(x)), x]).T 
```

```python
y = X @ np.array([0.5,2.0]) + np.random.normal(0,0.25,len(x))
```

```python
w = np.linalg.inv(X.T@X)@ X.T@y
```

```python
y_est = X @ w
```

```python
y_est
```

```python
y
```

```python
x
```

```python
np.linalg.norm(y-y_est)
```

```python
from sklearn import linear_model
import numpy as np 
x = np.random.random(10) 
X = np.vstack([np.ones(len(x)), x]).T
```

```python
print(x)
print(X)
```

```python
y = 0.5* np.ones(10) + 2.0*x + np.random.normal(0,0.25,len(x))
ols = linear_model.LinearRegression()
ols.fit(X,y)
y_est = ols.predict(X)
```

```python
print(y)
print(y_est)
```

```python
np.linalg.norm(y-y_est)
```
