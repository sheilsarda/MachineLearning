```python
# do a regression with two features x1 and x2 that are highly correlated
# What can we say about the "effect of" x1 and x2 on y?
import numpy as np
from sklearn import linear_model
x1 = np.random.random(300) 
x2 = 0.8*x1 + 0.2*np.random.random(300)
X = np.vstack((x1,x2)).T
y =  x1 - x2 + np.random.normal(0,0.1,len(x1))
print('corr(x1,x2):', np.corrcoef(x1,x2)[0,1])
print('corr(y,x1):', np.corrcoef(y,x1)[0,1])
print('corr(y,x2):', np.corrcoef(y,x2)[0,1])
ols = linear_model.LinearRegression()
ols.fit(X,y)
print('coeficients w1,w2:', ols.coef_)
```
